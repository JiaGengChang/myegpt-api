## CN 1q

```python
import pandas as pd
import dotenv
assert dotenv.load_dotenv('.env_eval')
from tools import convert_gene, _max_overlapping_segment
```

```python
cn_gene_names = ['BCL9', 'ANP32E', 'MCL1', 'RFX5', 'PSMD4', 'ILF2', 'IL6R', 'ADAR', 'CKS1B']
cn_gene_stable_ids = [convert_gene(gene) for gene in cn_gene_names]
cn_gene_dict = dict(zip(cn_gene_names, cn_gene_stable_ids))
cn_gene_dict
```

```python
cn_gene_segment_mean = {k : pd.Series(_max_overlapping_segment(v).query('sample.str.contains("_1_BM_CD138pos")')['segment_mean']) for k, v in cn_gene_dict.items()}
```

```python
df_cn_gene_segment_mean = pd.concat(cn_gene_segment_mean, axis=1)
```

```python
# sanity check to ensure no two columns (genes) have identical CN values
def equal_columns(df):
    cols = df.columns.tolist()
    pairs = []
    for i in range(len(cols)):
        for j in range(i + 1, len(cols)):
            c1, c2 = cols[i], cols[j]
            if df[c1].equals(df[c2]):
                pairs.append((c1, c2))
    return pairs

pairs = equal_columns(df_cn_gene_segment_mean)
print(pairs)
```

## TP53 survival

```python
import pandas as pd
import numpy as np
from lifelines import CoxPHFitter
```

```python
dfsurv = pd.read_csv('../clindata/MMRF_CoMMpass_IA22_STAND_ALONE_SURVIVAL.tsv', sep='\t')
```

```python
dfmut = pd.read_csv("../omicdata/IGV Downloads-MMRF_CoMMpass_IA22_exome_vcfmerger2_IGV_All_Canonical_NS_Variants.mut",sep='\t')
```

```python
px_tp53_mut = dfmut.loc[dfmut.GENE=='TP53']['sample'].str.extract(r'(MMRF_[0-9]+)_')[0].unique()
px_p53_wt = dfsurv.loc[~dfsurv.PUBLIC_ID.isin(px_tp53_mut),'PUBLIC_ID'].unique()
```

```python
dfcph = dfsurv[['PUBLIC_ID','oscdy','censos','pfscdy','censpfs']].copy()
dfcph = dfcph.merge(
    pd.DataFrame({'PUBLIC_ID': px_tp53_mut, 'tp53_status': True}),
    on='PUBLIC_ID',
    how='left'
)
dfcph['tp53_status'] = dfcph['tp53_status'].fillna(False)
```

```python
dfcphpfs = dfcph.drop(columns=['oscdy','censos']).dropna(subset=['pfscdy'])
dfcphos = dfcph.drop(columns=['pfscdy','censpfs'])
```

```python
cphos = CoxPHFitter()
cphos.fit(dfcphos.set_index('PUBLIC_ID'), duration_col='oscdy', event_col='censos')
cphos.print_summary()
```

```python
cphpfs = CoxPHFitter()
cphpfs.fit(dfcphpfs.set_index('PUBLIC_ID'), duration_col='pfscdy', event_col='censpfs')
cphpfs.print_summary()
```

```python
dfcphpfs_imp = dfcph.copy()
dfcphpfs_imp.loc[dfcph['pfscdy'].isna(),'pfscdy'] = dfcph.loc[dfcph['pfscdy'].isna(),'oscdy']
dfcphpfs_imp.drop(columns=['oscdy','censos'],inplace=True)
```

```python
cphpfs_imp = CoxPHFitter()
cphpfs_imp.fit(dfcphpfs_imp.set_index('PUBLIC_ID'), duration_col='pfscdy', event_col='censpfs')
cphpfs_imp.print_summary()
```
